HyperParameters

```python
testSize = 0.25
randomState = 4

n_estimators_start = 100
n_estimators_end = 150

max_depth_start = 2
max_depth_end = 20
```

```python
data = pd.read_csv("/content/drive/MyDrive/Workshop/DTree/Session3/train.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = testSize, random_state = randomState)
```

Find best values for n_estimators and max_depth

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
from tqdm.contrib import itertools



hyperParams = pd.DataFrame({"n_estimators":[],"max_depth":[],"accuracy":[]})



for estimator , depth in itertools.product(range(n_estimators_start , n_estimators_end + 1) ,range(max_depth_start , max_depth_end + 1)):
    clf = RandomForestClassifier( n_estimators = estimator, max_depth = depth)
    clf.fit(X_train, y_train)

    y_pred=clf.predict(X_test)

    accuracy = metrics.accuracy_score(y_test,y_pred)

    new_hp = pd.DataFrame({"n_estimators":[estimator],"max_depth":[depth],"accuracy":[accuracy]})

    hyperParams = hyperParams.append(new_hp)

```

```python
result = hyperParams['accuracy'].values
best_HP = hyperParams.iloc[result.argmax(),:]

best_n_estimators = best_HP['n_estimators']
best_max_depth = best_HP['max_depth']

print( int(best_n_estimators))
print( int(best_max_depth))
print('Best Hyper Parameters is:')
print(best_HP)
```

# **Create Best Model**

```python
clf = RandomForestClassifier( n_estimators = int(best_n_estimators), max_depth = int(best_max_depth))
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)

accuracy = metrics.accuracy_score(y_test,y_pred)
accuracy
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
